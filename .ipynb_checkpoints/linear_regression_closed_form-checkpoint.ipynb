```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
import seaborn as sns
```

```python
sns.set()
```

```python
# Generate Dataset
x,y = make_regression(n_samples=300,n_features=1,n_informative=1,noise=1.8,random_state=11)


y = y.reshape((-1,1))
print(x.shape)
print(y.shape)
```

```python
# Normalize
X = (x-x.mean())/x.std()
Y = y
# Visualize
plt.figure()
plt.scatter(X,Y)
plt.title("Normalized Data")
plt.show()
```

```python
plt.scatter(X,Y,label="Normalized Data")
plt.scatter(x,y,color="orange",label="Original Data")
plt.legend()
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
X_ = np.hstack((X,ones))
print(X_.shape)
print(X_[:5,:])
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def getThetaClosedForm(X,Y):
    
    Y = np.mat(Y)
    
    theta = np.dot(np.linalg.pinv(np.dot(X.T,X)),(np.dot(X.T,Y)))
    return theta
```

```python
theta = getThetaClosedForm(X_,Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color='red',label="prediction")
plt.title("Normalized Data")
plt.legend()
plt.show()
```
