```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from jupyterthemes import jtplot
jtplot.style()
```

```python
arr = np.random.randn(1000)
```

```python
print(arr.shape)
```

```python
plt.hist(arr,bins=10)
plt.xlabel("X")
plt.ylabel("Prob(X)")
plt.show()
```

```python
np.mean(arr)
```

```python
np.std(arr)
```

```python
marks = np.random.randn(1000)*10 + 60
```

```python
np.mean(marks)
```

```python
np.std(marks)
```

## Standard Normal Distribution and Normal Distribution

```python
plt.hist(arr,bins=10,label="Standard Normal Distribution")
plt.hist(marks,bins=10,label="Normal Distribution")
plt.legend()
plt.show()
```
