```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.asarray([1,2,3,4,5])
y = np.asarray(x+1)
```

```python
plt.plot(x,y)
plt.show()
plt.scatter(x,y, color="orange")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot")
```

```python
plt.figure()
plt.plot(x,y,color="orange", label="linear")
plt.plot(x,y**2,color="red", label="polynomial")
plt.xlabel("X-Axis")
plt.ylabel("Price")
plt.legend()
plt.title("Scatter plot")
plt.show()
```
