```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from jupyterthemes import jtplot
jtplot.style()
```

```python
dfx = pd.read_csv('Dataset/weightedX.csv')
dfy = pd.read_csv('Dataset/weightedY.csv')
```

```python
X = dfx.values
Y = dfy.values
```

```python
print(X.shape)
print(Y.shape)
```

```python
u = X.mean()
std = X.std()
X = (X-u)/std
plt.title("Data Normalized")
plt.scatter(X,Y)
plt.show()
```

```python
def getW(X_,X,tau):
    M = X.shape[0]
    W = np.mat(np.eye(M))
    
    for i in range(M):
        xi = X[i]
        x  = X_
        W[i,i] =  np.exp(np.dot((xi-x),(xi-x).T)/(-2*tau*tau))
    return W
```

```python
X = np.mat(X)
Y = np.mat(Y)
M = X.shape[0]

W = getW(-1,X,100)
print(W.shape)
print(W)
```

```python
def predict(X,Y,query_x,tau):
    ones = np.ones((M,1))
    X_ = np.hstack((X,ones))
    
    qx = np.mat([query_x,1])
    
    W = getW(qx,X_,tau)
    
    #theta = `(X′WX)inv * X′WY`
    theta = np.linalg.pinv(X_.T*(W*X_))*(X_.T*(W*Y))
    #print(theta.shape)
    pred = np.dot(qx,theta)
    return theta,pred
```

```python
theta,pred = predict(X,Y,1.0,1.0)
```

```python
print(theta)
```

```python
print(pred)
```

```python
def plotPrediction(tau):
    X_test = np.linspace(-2,2,20)
    Y_test = []
    
    for xq in X_test:
        theta,pred = predict(X,Y,xq,tau)
        Y_test.append(pred[0][0])
        
    Y_test = np.array(Y_test)
    
    XO = np.array(X)
    YO = np.array(Y)
    
    plt.title("Tau/Bandwidth Param %.2f"%tau)
    plt.scatter(XO,YO)
    plt.scatter(X_test,Y_test,color='red')
    plt.show()
```

```python
taus = [0.1,0.5,1,5,10]
for t in taus:
    plotPrediction(t)
```
