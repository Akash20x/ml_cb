## Gradient Descent Algorithm

```python
x1 = np.random.randint(10)
print(x1)
```

```python
steps = 1
learing_rate = 0.01
while (steps<=30):
    x1 -= learing_rate*2*(x1-5)
    
    error = ((x1-5)**2)
    plt.scatter(x1,error)
#     print("X %.4f ,Error %.4f "%(x,error))
    steps += 1

    (x1,(x1-5)**2)
```
