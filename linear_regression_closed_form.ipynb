```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
import seaborn as sns
```

```python
sns.set()
```

```python
# Generate Dataset
x,y = make_regression(n_samples=300,n_features=1,n_informative=1,noise=1.8,random_state=11)


y = y.reshape((-1,1))
print(x.shape)
print(y.shape)
```

```python
# Normalize
X = (x-x.mean())/x.std()
Y = y
# Visualize
plt.figure()
plt.scatter(X,Y)
plt.title("Normalized Data")
plt.show()
```

```python
plt.scatter(X,Y,label="Normalized Data")
plt.scatter(x,y,color="orange",label="Original Data")
plt.legend()
plt.show()
```
